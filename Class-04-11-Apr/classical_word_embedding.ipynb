### average wrod2vec

sentence = "I love machine leaning"

sentence -> vector? 

average word2vec concept

first -> create a vector for each word in the sentence

then -> take the average of all the vectors to get a single vector representation for the entire sentence


```python
sentence = "I love machine leaning"
```

```python
sentence.lower().split()
```

```python
words=sentence.lower().split()
```

```python
word_vectors = []
```

```python
for word in words:
    print(word)
    word_vectors.append(model[word])
```

```python
word_vectors
```

```python
import numpy as np
sentence_vector = np.mean(word_vectors,axis=0)
sentence_vector
```

```python
sentence_vector.shape
```

word2vec -> every word into the KeyedVectors
