### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
documents = ["people watch movie", 
             "people watch cricket", 
             "people like movie", 
             "people like cricket"]
```

```python
tokens = [sentence.lower().split() for sentence in documents]
```

```python
tokens
```

```python
all_words = [[word] for sentence in tokens for word in sentence]
```

```python
all_words
```

```python
type(all_words)
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

#### sparse_output=False means the encoder returns a NumPy array instead of a sparse matrix. This makes the output easier
https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

```python
encoder.fit(all_words)
```

```python
print("Vocabulary:", encoder.categories_[0])
```

```python
tokens
```

```python
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print( "Encoded:\n", encoded_sentence)
```

### Bag of Words
https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html#sklearn.feature_extraction.text.CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket", 
             "people like movie and like movie a lot", 
             "people like cricket"]
```

### AIM: Covert Data into Number

```python
bow = CountVectorizer()
```

```python
bow_matrix = bow.fit_transform(documents)
```

```python
bow_matrix
```

```python
print("BOW Matrix:\n", bow_matrix.toarray())
```

```python
print('Vocabulary:', bow.get_feature_names_out())
```

```python
bow.fit(documents)
```

```python
documents
```

```python
bow.transform(documents).toarray()
```

```python
document=["lion is the king of the jungle"]
```

```python
bow.transform(document).toarray()
```

### TF-IDF Practical

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket", 
             "people like movie and like movie a lot", 
             "people like cricket"]
```

```python
tf_idf = TfidfVectorizer()
```

```python
tf_idf_vector=tf_idf.fit_transform(documents)
```

```python
tf_idf_vector
```

```python
tf_idf_vector.toarray()
```

```python
for vector in tf_idf_vector.toarray():
    print("\nTF-IDF Vector:\n", vector)
```

```python
print("Vocabulary:", tf_idf.get_feature_names_out())
```
